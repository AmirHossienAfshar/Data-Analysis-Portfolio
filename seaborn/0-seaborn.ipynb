### scatter plots
As mentioned above, a scatter plot is used to visualize the relationship between two variables. Here is what it's neccesery parameters look like:
- `x` and `y` correspond to the columns from the DataFrame used to plot the data on the x-axis and y-axis.
- `hue` adds a categorical variable that colors the points differently based on its values.
- `size` controls the size of the points based on another variable in the DataFrame.
- `sizes` controls the minimum and maximum size of the points.
- `data` is the DataFrame that contains the data.

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", size="size", sizes=(50, 200), data=tips)

# add labels and title
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip")

# display the plot
plt.show()
```

#### why is the code above seems to be a little bit wierd?
Pyplot is a **state-based** interface to a Matplotlib module that provides MATLAB-like functionality. what is that mean?
Both Seaborn and Matplotlib work with the concept of a "**current figure**" or "**current axes.**" You don't need to assign the plot to a variable because these libraries keep track of the active figure/plot automatically. This shared state allows you to modify the plot without needing to refer to it explicitly.


The concept of current figure and current axes in Matplotlib allows for the creation and modification of **one plot at a time** unless explicitly managed. 
If you wish to know more about that, checkout [here](https://stackoverflow.com/questions/38415774/ipython-get-access-to-current-figure), and learn more about `plt.gcf()` and `plt.gca()`.

#### line plots
In Seaborn, line plots are used to visualize data where the relationship between variables is continuous, typically involving time series or other sequences. 
##### Key Parameters of sns.lineplot()
- `x`: The data for the x-axis, which is usually continuous.
- `y`: The data for the y-axis, representing the dependent variable.
- `hue`: Grouping variable that will produce lines with different colors.
- `style`: A categorical variable that changes the style of the lines (e.g., solid, dashed). Example: style="event" changes line styles based on a column.
- `size`: Variable that determines the thickness of the line. Example: size="signal" would vary the width of the lines based on signal values.

```python
tips.head()
```

```python
sns.lineplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip")
plt.show()
```

### Seaborn bar plots
Bar plots are used to visualize the relationship between a categorical variable and a continuous variable. In a bar plot, each bar represents the mean or median (or any aggregation) of the continuous variable for each category. In Seaborn, bar plots can be created using the barplot() function. 
#### Key Parameters of sns.barplot()
- `x`: This represents the categorical variable to be plotted along the x-axis.
- `y`: The numeric variable you want to summarize, which will be plotted along the y-axis.
- `hue`: A categorical variable to further split the data and plot bars with different colors.
- `estimator`: A function that defines how the data is aggregated. By default, it’s the mean, but it can be changed to functions like np.median, np.sum, etc.
